```python
#import
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import chisquare
from scipy import stats
```

```python
def generate(L):
    normal = norm.rvs(-6, 2, L)
    hist, bins = np.histogram(normal, bins = 25, density = True)
    plt.hist(normal, 25, density=True)
    plt.title('L = ' + str(L))
      
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, L)
    tpdf = norm.pdf(x, -6, 2)
    
    plt.plot(x, tpdf, 'k')
    plt.show()
    
    tpdf,bins2 = np.histogram(tpdf, bins = 25)
    
    tstat = 0
    for i in range(len(hist)):
        tstat = tstat + ((hist[i] - tpdf[i])**2 / tpdf[i])/L
    
    true_tstat = 1 - tstat

    print(true_tstat)
```

```python
generate(10**2)
generate(10**3)
generate(10**4)
```
